```python
#Reading in the data
import pandas as pd 
import math
import requests

nvidia_frames_data = pd.read_csv('nvidia_frames.csv') 
lidar_frames_data = pd.read_csv('lidar_frames.csv')
```

```python
#Droping irrelevant columns from dataframe
nvidia_frames_data.drop(columns=['yaw', 'roll', 'pitch', 'left_filename', 'right_filename'], inplace=True)

#Removing parts where car doesn't move
data = nvidia_frames_data[nvidia_frames_data['vehicle_speed'] > 0]

#Removing parts where car has turn signal on
data = data[data['turn_signal'] == 1]

#Removing NaN value rows from dataset
data.dropna()
```

```python
#Replacing koordinates in the dataset - Takes about 15 minutes, so instead you can import and use already converted dataset
#in the below cell
for i in range(31496):
    j = 6
    k = 7
    x = data.iloc[i].position_x + 650000
    y = data.iloc[i].position_y + 6465000
    URL = "https://gpa.maaamet.ee/calc/geo-lest/url/?xy=" + str(x) + "," + str(y)
    page = requests.get(URL)
    koordinaadid = page.text.strip().split(',')
    data.iloc[i,j] = koordinaadid[0]
    data.iloc[i,k] = koordinaadid[1]
    
data.to_csv('nvidia_frames_Converted.csv')
```

```python
data = pd.read_csv('nvidia_frames_Converted.csv')
data = data.iloc[: , 1:]
```

```python
#coverting steering angle from radian to degrees
for i in range(data.shape[0]):
    j = 3
    data.iloc[i,j] = abs(math.degrees(data.iloc[i,j]))
    
```

```python
data.drop(columns=['position_z'], inplace=True)
data['position_x'] = data['position_x'].astype(float)
data['position_y'] = data['position_y'].astype(float)

#Retrieving minimal turn radius to be considered a turn
df = pd.DataFrame()
df = data

#converting speed to km/h
df['vehicle_speed'] = df['vehicle_speed'].replace([df['vehicle_speed']],[df['vehicle_speed']*3.6])


print('Maximum steering angle' + str(df['steering_angle'].max()))
print('Minimum steering angle' + str(df['steering_angle'].min()))
```

```python
#converting index/datetime to seconds in order to select few seconds before the curve

#             Hetkel ei ole kasutuses

index = data['index']

for i in range(index.shape[0]):
    
    index.iloc[i] = index.iloc[i].split('\n')
    
    index.iloc[i] = index.iloc[i][0][12:19]
    
    index.iloc[i] = index.iloc[i].split(':')
    
    sekundid = 0
    aeg = index.iloc[i]
    for j in range(len(aeg)):
        if j == 0:
            aeg[j] = int(aeg[j])
            sekundid += aeg[j]*3600
        elif j == 1:
            aeg[j] = int(aeg[j])
            sekundid += aeg[j]*60
        else:
            aeg[j] = int(aeg[j])
            sekundid += aeg[j]
    
    index.iloc[i] = sekundid
```

```python
#seperating the dataframe into two - straight parts of the road and curved parts
df_kurvid = pd.DataFrame()
data_alles = pd.DataFrame()
index = 0
while index < data.shape[0]:
    
    if float(data.iloc[index, 3]) >= 36:
        for j in range(35):
            df1 = data.iloc[index-35+j]
            df_kurvid = df_kurvid.append(df1, ignore_index=True)
            data.iloc[index-35+j, 2] = 'rm'
            
        
        while float(data.iloc[index, 3]) >= 36:
            for k in range(35):  
                df_kurvid = df_kurvid.append(data.iloc[index+k], ignore_index=True)
                data.iloc[index+k, 2] = 'rm'
                if data.shape[0] < index+36:
                    break
            if data.shape[0] < index+36:
                break    
            
            index += 35
    
    index += 1
    
    if data.shape[0] < index+36:
        data_alles = data.iloc[index:]
        for a in range(data_alles.shape[0]):
            if float(data_alles.iloc[a,3]) >= 36:
                df_kurvid = df_kurvid.append(data_alles.iloc[a], ignore_index=True)
        
        break    
df_sirged = data[data['autonomous'] != 'rm']
```

```python
#Splitting the intersections coordinates
inter_data = pd.read_csv('intersections_coordinates.csv')
inter_data
inter_data['start_x'] = 0
inter_data['start_y'] = 0
inter_data['end_x'] = 0
inter_data['end_y'] = 0
inter_data
```

```python
for i in range(inter_data.shape[0]):
    inter_data['start'][i] = inter_data['start'][i][1:-1].split(',')
    inter_data['start_x'].iloc[i] = float(inter_data['start'][i][0])
    inter_data['start_y'].iloc[i] = float(inter_data['start'][i][1])
    inter_data['end'][i] = inter_data['end'][i][1:-1].split(',')
    inter_data['end_x'].iloc[i] = float(inter_data['end'][i][0])
    inter_data['end_y'].iloc[i] = float(inter_data['end'][i][1])
    
inter_data
inter_data.drop(columns=['start', 'end'])
```

```python
def f(x, y):
    boolean = False
    for i in range(inter_data.shape[0]):
        if (inter_data['start_x'].iloc[i] > x) & (x > inter_data['end_x'].iloc[i]) & (inter_data['start_y'].iloc[i] > y) & (y > inter_data['end_y'].iloc[i]):
            boolean = True
        elif (inter_data['start_x'].iloc[i] < x) & (x < inter_data['end_x'].iloc[i]) & (inter_data['start_y'].iloc[i] < y) & (y < inter_data['end_y'].iloc[i]):
            boolean = True 
        elif (inter_data['start_x'].iloc[i] > x) & (x > inter_data['end_x'].iloc[i]) & (inter_data['start_y'].iloc[i] < y) & (y < inter_data['end_y'].iloc[i]):
            boolean = True
        elif (inter_data['start_x'].iloc[i] < x) & (x < inter_data['end_x'].iloc[i]) & (inter_data['start_y'].iloc[i] > y) & (y > inter_data['end_y'].iloc[i]):
            boolean = True
    return boolean
```

```python
non_intersections_data = pd.DataFrame(columns=data.columns)
intersections_data = pd.DataFrame(columns=data.columns)
    
for i in range(data.shape[0]):
    on_ristmik = f(data['position_x'].iloc[i], data['position_y'].iloc[i])
    if on_ristmik:
        intersections_data = intersections_data.append(data.iloc[i], ignore_index=True)
        
    else:
        non_intersections_data = non_intersections_data.append(data.iloc[i], ignore_index=True)
        
intersections_data
```
